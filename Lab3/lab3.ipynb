```python
import numpy as np
import sklearn
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import metrics

#1
dataset = make_regression(2800000, n_features=1, noise=3)
#print(dataset)
model = LinearRegression()
model.fit(dataset[0], dataset[1])
y_pred = model.predict(dataset[0])

plt.plot(dataset[0],dataset[1], 'rd', dataset[0], y_pred, '--k')

print('Mean Absolute Error:', metrics.mean_absolute_error(dataset[1], y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(dataset[1], y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(dataset[1], y_pred)))
```

```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X, y = datasets.make_classification(n_samples=300, n_features=5, n_classes=2, n_redundant=0,
                                    n_clusters_per_class=1, random_state=42)

fig, ax = plt.subplots(figsize=(9, 6))
plt.xlabel("X0", fontsize=20)
plt.ylabel("X1", fontsize=20)
plt.scatter(X[:,0], X[:,1], s=50, c=y)

# Training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
pipeline = make_pipeline(StandardScaler(), LogisticRegression())
pipeline.fit(X_train, y_train)
print("Model score is ", pipeline.score(X_test, y_test))
x = np.array([1, 0.5, 0.4, 0.3, 0.2]).reshape(1, -1)
print(pipeline.predict(x))
x = np.array([0, 0, 0, 0, 0]).reshape(1, -1)
print(pipeline.predict(x))
```
